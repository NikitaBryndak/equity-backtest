```python
import pandas as pd
import re
```

```python
df = pd.read_csv('../data/data.csv')
```

```python
df.head()
```

```python
volume_cols = [c for c in df.columns if re.search('volume', c, flags=re.IGNORECASE)]
ohlc = [c for c in ['Open', 'High', 'Low', 'Close'] if c in df.columns]
columns_to_use = ohlc + [c for c in volume_cols if c not in ohlc]
df = df[columns_to_use]
df
```

```python
import yfinance as yf

ticker = yf.Ticker("MSFT")
ticker.history(start="2015-01-01", end="2025-12-31", interval="1d")
ticker.history(period="max", interval="1d").to_csv('../data/MSFT.csv')
```
