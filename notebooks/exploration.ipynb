```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.set_index('Date', inplace=True)
df.sort_index(inplace=True)
```

```python
df.drop(['Symbol', 'Timestamp'], axis=1, inplace=True)
df
```

```python
df['SMA_30'] = df['Close'].rolling(window=30).mean()
df['SMA_100'] = df['Close'].rolling(window=100).mean()
```

```python
df.tail(100)
```

```python
df['Signal'] = 0
df.loc[df['Close'] < df['SMA_100'], 'Signal'] = 1
df.loc[df['Close'] > df['SMA_30'], 'Signal'] = -1
df.tail(100)
```

```python
df['Signal'].value_counts()
```

```python
df['Returns'] = df['Close'].pct_change().fillna(0)
df
```

```python
df['Strategy_Returns'] = (df['Returns'] * df['Signal'].shift(1)).fillna(0)
df['Strategy_Equity'] = (1 + df['Strategy_Returns']).cumprod()
df['BuyHold_Equity'] = (1 + df['Returns']).cumprod()
```

```python
plt.figure(figsize=(12, 6))
df['Strategy_Equity'].plot(label='Strategy')
df['BuyHold_Equity'].plot(label='Buy & Hold')
plt.title('Strategy vs. Buy & Hold Equity Curves')
plt.ylabel('Equity')
plt.xlabel('Date')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df[['Returns', 'Signal', 'Strategy_Returns', 'Strategy_Equity', 'BuyHold_Equity']].tail(10)
```

```python
subset = df.loc['2017-06-01':'2022-01-01']
plt.figure(figsize=(12, 6))
subset['Strategy_Equity'].plot(label='Strategy')
subset['BuyHold_Equity'].plot(label='Buy & Hold')
plt.title('Strategy vs. Buy & Hold Equity Curves')
plt.ylabel('Equity')
plt.xlabel('Date')
plt.legend()
plt.grid(True)
plt.show()
```

```python
daily_df = df.resample('1D').agg({
    'Open': 'first',
    'High': 'max',
    'Low': 'min',
    'Close': 'last',
    'Volume BTC': 'sum', 
    'Volume USD': 'sum'
}).dropna()

daily_df.head()
```

```python
daily_df['Returns'] = daily_df['Close'].pct_change()
```

```python
daily_df['SMA_30'] = daily_df['Close'].rolling(30).mean()
daily_df['SMA_100'] = daily_df['Close'].rolling(100).mean()

daily_df['Signal'] = 0
daily_df.loc[daily_df['Close'] < daily_df['SMA_100'], 'Signal'] = 1
daily_df.loc[daily_df['Close'] > daily_df['SMA_30'], 'Signal'] = -1
```

```python
daily_df['Strategy_Returns'] = daily_df['Returns'] * daily_df['Signal'].shift(1)
```

```python
daily_df['Strategy_Equity'] = (1 + daily_df['Strategy_Returns']).cumprod()
daily_df['BuyHold_Equity'] = (1 + daily_df['Returns']).cumprod()

plt.figure(figsize=(12, 6))
daily_df['Strategy_Equity'].plot(label='Strategy')
daily_df['BuyHold_Equity'].plot(label='Buy & Hold')
plt.title('Daily SMA Strategy vs Buy & Hold')
plt.legend()
plt.grid(True)
plt.show()
```

```python
daily_df[['Returns', 'Signal', 'Strategy_Returns', 'Strategy_Equity', 'BuyHold_Equity']].tail(10)
```

```python
daily_df['Signal'] = np.where(daily_df['SMA_30'] > daily_df['SMA_100'], 1, -1)
```

```python
daily_df['Strategy_Returns'] = daily_df['Returns'].shift(-1) * daily_df['Signal']
```

```python
daily_df['Strategy_Equity'] = (1 + daily_df['Strategy_Returns']).cumprod()
daily_df['BuyHold_Equity'] = (1 + daily_df['Returns']).cumprod()

plt.figure(figsize=(12, 6))
daily_df['Strategy_Equity'].plot(label='Strategy')
daily_df['BuyHold_Equity'].plot(label='Buy & Hold')
plt.title('Daily SMA(trend following) Strategy vs Buy & Hold')
plt.legend()
plt.grid(True)
plt.show()
```
